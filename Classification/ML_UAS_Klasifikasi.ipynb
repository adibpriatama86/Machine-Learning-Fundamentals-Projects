```python
import pandas as pd
import numpy as np

#load dataset
df = pd.read_csv("/content/accident.csv")

#cek missing value
print(df.isnull().sum())
# drop mssing value
df = df.dropna()

df = df.reset_index(drop=True)

#Gender: Male = 0, Female = 1
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

#Helmet_Used: No = 0, Yes = 1
df['Helmet_Used'] = df['Helmet_Used'].map({'No': 0, 'Yes': 1})

#Seatbelt_Used: No = 0,Yes = 1
df['Seatbelt_Used'] = df['Seatbelt_Used'].map({'No': 0, 'Yes': 1})

print(df.head())
print(df.info())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

#pisahkan fitur dan target
X = df[['Age', 'Gender', 'Speed_of_Impact', 'Helmet_Used', 'Seatbelt_Used']]
y = df['Survived']

#splitting data
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    random_state=42
)

#model NN
clf = MLPClassifier(
    hidden_layer_sizes=(20, 30),
    solver='lbfgs',
    alpha=1e-5,
    random_state=1
)

#training model
clf.fit(X_train, y_train)

#prediksi
y_pred = clf.predict(X_test)

#evaluasi performa
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Hasil evaluasi model:")
print(f"Akurasi: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

#confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)
```
